## Applied Pandas Example 

Here we show an example of using the Python library Pandas to deal with some sampled weather data. 

### Importing Pandas and other libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Creating some random meteorological data: 

```python
data = {
    "Timestamp": pd.date_range(start="2025-01-01", periods=48, freq="H"),
    "Temperature_C": np.random.normal(loc=15, scale=5, size=48),
    "Humidity_%": np.random.uniform(40, 90, size=48),
    "Wind_Speed_mps": np.random.uniform(0, 15, size=48),
}
df = pd.DataFrame(data)
df.set_index("Timestamp", inplace=True)
```

### Printing some basic statistics: 

```python
print("Basic Statistics:")
print(df.describe())
```

### We can also plot a time-series of our data: 

```python
df.plot(subplots=True, figsize=(10, 6), title="Time Series")
plt.show()
```

### We can also find temperature anomalies within our data: 

```python
mean_temp = df["Temperature_C"].mean()
std_temp = df["Temperature_C"].std()
df["Temp_Anomaly"] = np.abs(df["Temperature_C"] - mean_temp) > 2 * std_temp
print("\nTemperature Anomalies:")
print(df[df["Temp_Anomaly"]])
```
